```python
import requests
from openpyxl import load_workbook
import pandas as pd 
import os
```

```python
def saveExcelfromURL(url, fileName, path, sheetName, skipRow):
    global df    
    resp = requests.get(url)
    output = open(fileName, 'wb')
    output.write(resp.content)
    output.close()
    wb = load_workbook(fileName)
    filePath = (path + fileName)
    
    data = pd.read_excel(filePath, sheetName, skiprows=skipRow, index_col=None, na_values=['NA'])
    df = pd.DataFrame(data)
```

```python
def vlookup(df1, df2, matchID, varLookup):
    df1 = df1[[matchID, varLookup]]
    return pd.merge(df2, df1, how='left', on=matchID)
```
